Converter para segundos

```
def convert_hora(x):
  if(len(x)==1):
    return "00:0"+x
  elif(len(x)==2):
    return "00:"+x
  elif(len(x)==3):
    return "0"+x[0]+":"+x[1:]
  elif(len(x)==4):
    return x[0:2]+":"+x[2:]


```
